```python
x = pd.Series([1, np.nan, None, 5, pd.NA, 'hello'])
print(x)
print(x.isnull())
print(x.notnull())
print(x[x.isnull()])
print(x[x.notnull()])

print(x.dropna())
```

```python
x = pd.DataFrame([[1, np.nan, None], [1, 2, 3], [5, pd.NA, 3]])
print(x)
print(x.dropna(axis=0))
print(x.dropna(axis=1))

print(x.dropna(axis=0, how='all'))

x = pd.DataFrame([[None, np.nan, None], [1, 2, 3], [5, pd.NA, 3]])
print(x.dropna(how='all'))
# how = all - убрать если ВСЕ значения отсутсвуют
# how = any - убрать если все хотя бы ОДНО значения отсутсвуют

x = pd.DataFrame([[None, np.nan, None], [1, 2, 3], [2, np.nan, 3]])
print(x)
print(x.dropna(axis=0, thresh=0))  # Должно быть МИНИМУМ N непустых значений
```

```python
x = pd.DataFrame([None, np.nan,3,  None, 1, 2, 3], dtype="Int32")
print(x)
print(x.fillna(-4))
print(x.ffill()) # заполнение предыдущим
print(x.bfill())# заполнение следующим
```

```python
x = pd.DataFrame([[1, np.nan, None], [1, 2, 3], [2, np.nan, 3]])
print(x)
print(x.fillna(-4))
print(x.ffill(axis=1))  # заполнение предыдущим значением
print(x.bfill(axis=1))  # заполнение след значением
print(x.ffill(axis=0))  # заполнение предыдущим значением
print(x.bfill(axis=0))  # заполнение след значением
```

```python
# Иерархическая индексация

index = [
    ("A1", 2025),
    ("A1", 2026),
    ("A2", 2025),
    ("A2", 2026),
    ("A3", 2025),
    ("A3", 2026)]
data = [1, 2, 3, 4, 5, 6]
s = pd.Series(data, index=index)
print(s)

# print(s[[i for i in s.index if i[1] == 2025]])
mi = pd.MultiIndex.from_tuples(index)
print(mi)

s = s.reindex(mi)
print(s)

s1 = pd.Series(data, index=mi)
print(s1)

print(s['A1',:])

df = s.unstack()
print(df)

print(df.stack()) # Одномерные Series может хранить данные с большим числом измерений
```
