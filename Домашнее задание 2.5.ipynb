```python
import numpy as np
import pandas as pd
```

```python
data = pd.Series([0.25, 0.5, 0.75, 1.0])
print(data)
print(data.values)
print(data.index)
print(type(data.index))

print(data[0])
print(data[1:3])

data = pd.Series([0.25, 0.5, 0.75, 1.0], index=[1, 10, 7, 'd'])
print(data)
print(type(data.index[0]))
print(type(data.index[3]))

print(data[10: 'd'])
```

```python
dict = {'A': 10, 'B': 20, 'C': 30, 'D': 40, 'E': 50}
data_dict = pd.Series(dict)
print(data_dict)
print(data_dict['B'])
print(data_dict['B':'D'])

# списки
data = pd.Series([0.25, 0.5, 0.75, 1.0])
print(data)

data = pd.Series(5, index=[10, 20, 30])
print(data)

# словарь ключи в метки поедут
data_dict = pd.Series(dict, index=['A', 'D'])
print(data_dict)

dict1 = {"A": 10, "B": 20, "C": 30, "D": 40, "E": 50}
data1 = pd.Series(dict1)
print(data)

dict2 = {"A": 11, "B": 21, "C": 31, "D": 41, "E": 51}
data2 = pd.Series(dict2)
print(data)

df = pd.DataFrame({1: data1, 2: data2})
print(df)
print(df.index)
print(type(df.index))
print(df.columns)
print(type(df.columns))

print(df[1])
print(df[2])

df = pd.DataFrame({1: data1})
print(df)
print(df.values)

df = pd.DataFrame(data1, columns=['rrr'])
print(df)

print(pd.DataFrame([{"a": i, "b": 2 * i} for i in range(4)]))
# Nan - Not a Number
print(pd.DataFrame(np.zeros((3, 4))))
```

```python
# INDEX - способ сослаться на данные в сериес или датафрейм
index = pd.Index([2,5,3,5,71])
print(index)
print(type(index))

print(index[1])
print(index[::2])

# индексы менять нельзя
```

```python
index1 = pd.Index([2,5,3,5,71])
index2 = pd.Index([1,2,51,4,99])

print(index1.intersection(index2))
print(index1.union(index2))
```

```python
# NumPy
# индексация arr[1, 2]
# срезы arr[::]
# маскирование arr[arr>5]
# arr[0, [1, 5]], arr[:, [1, 5]]

data = pd.Series([0.25, 0.5, 0.75, 1], index=["a", "b", "c", "d"])
print(data)

print(data["a"])
print("a" in data)
print("a1" in data)

print(list(data.items()))

print(data)
data["a"] = 99

data["a1"] = 999
print(data)

print(data["c":"a1"])

print(data[2:4])
print(data[2:])

print(data[(data > 0.3) & (data < 0.8)])

print(data.loc[["c", "a"]])

print(data.iloc[[2, 0]])
```
