```python
# Работа с временными рядами
# конкретные моменты времени, т.е. абсолютные значения "1 января 2026 года 00:00"
# периоды или временные интервалы, "март 2025 года", "24 часа"
# продолжительность(временная дельта) = 15 мин

# Python
from datetime import datetime
import dateutil

d = datetime(year=2026, month=3, day=4)
print(d)
print(type(d))

d = dateutil.parser.parse('4th of March, 2026')
print(d)
print(type(d))

print(d.strftime('%A'))
```

```python
# Numpy
print(d + 1)
d1 = np.array('2026-03-04 00:00', dtype=np.datetime64)
print(d1)
print(d1.dtype)
print(d1 + 1)

d = np.datetime64('2026-03-04', "D")
print(d)
print(d + 1)
d = np.datetime64('2026-03-04', "Y")
print(d)
print(d + 1)

d = np.datetime64('2026-03-04', "W")
print(d)
print(d + 1)

d = np.datetime64('2026-03-04', "ns")
print(d)
print(d + 1)

t = np.timedelta64(1, "D")
print(t)
print(t + 1)

t1 = np.timedelta64(1, "M")
print(t1)
print(t1 + 1)

# print(t1 + t) нельзя сложить по разному базису, юниты должны быть совместимыми
```

```python
# Pandas
d = pd.to_datetime('4th of March, 2026')
print(d)
print(type(d))
print(d.strftime('%A'))

d2 = d + pd.to_timedelta(np.arange(30), "D")
print(d2)

index = d2
data = pd.Series(np.arange(30), index=index)
print(data)

print(data['2026-03-04':"2026-03-10"])

print(data["2026-04"])
```

```python
# Работа с временными рядами
# конкретные моменты времени, т.е. абсолютные значения "1 января 2026 года 00:00" -
# Pandas timestamp(numpy.datetime64)
# периоды или временные интервалы, "март 2025 года", "24 часа" -
# pandas: Period (numpy.datetime64) - PeriodIndex
# продолжительность(временная дельта) = 15 мин -
# pandas timedelta (numpy.timedelta64) - TimedeltaIndex
d = pd.to_datetime('4th of March, 2026')
print(d)
print(type(d))

ds = pd.to_datetime(['2026-03-04', '2026-03-05'])
print(ds)
p = ds.to_period("D")
print(p)
print(type(p))

delt = ds[1] - ds[0]
print(delt)
print(type(delt))

idelt = ds - ds[0]
print(idelt)
print(type(idelt))

print(pd.date_range(start='2026-01-01', end='2026-03-04'))

print(pd.date_range(start='2026-01-01', periods=10, freq='h'))

hh = pd.date_range(start='2026-01-01', periods=10, freq='h')
print(hh[0]+2*hh.freq)

print(pd.period_range(start='2026-01-01', periods=10, freq='M'))

print(pd.timedelta_range(0, periods=10, freq='h'))
```

```python
# Коды периодичности

# M Q A - конец
# MS QS AS - конец

p1 = pd.Period("2026Q1")
print(p1)
print(p1.month)
print(p1.day)

print(pd.timedelta_range(0, periods=10, freq="2h15min"))

from pandas.tseries.offsets import BDay
x = pd.date_range("2026-01-01", periods=10, freq=BDay())
print(x)
```
