```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.externals.array_api_compat.cupy import linspace
```

```python
def f(x,y):
    return np.sin(np.sqrt(x**2 + y**2))

x = np.linspace(-6,6,30)
y = np.linspace(-10,10,50)

X,Y = np.meshgrid(x,y)
Z = f(X,Y)

fig = plt.figure()
ax = plt.axes(projection='3d')

# ax.scatter3D(X, Y, Z, c=Z)
# ax.plot_wireframe(X, Y, Z)

ax.plot_surface(X, Y, Z, cmap='rainbow')

plt.show()
```

```python
def f(x, y):
    return np.sin(np.sqrt(x ** 2 + y ** 2))

angle = np.linspace(0, 2*np.pi, 50)
r = np.linspace(0, 6, 30)

R, Angle = np.meshgrid(r, angle)

X = R * np.sin(Angle)
Y = R * np.cos(Angle)
Z = f(X, Y)

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, Z, cmap='rainbow')
plt.show()
```

```python
#seaborn
import seaborn as sns
```

```python
sns.set_style("darkgrid")

cars = pd.read_csv('/Users/adil.murat/PycharmProjects/ml_formula/data/cars.csv.gz')
print(cars.head())

## датасеты включают в себя числа(год например) и/или категории(вид двигателя, КПП)

# числовые данные
# парная диаграмма
sns.pairplot(data=cars, hue='transmission')
plt.show()
plt.show()
```

```python
# тепловая карта
cars_corr = cars[['year', 'selling_price', 'seats', 'mileage']]
sns.heatmap(cars_corr.corr(), cmap='viridis', annot=True)

plt.show()
```

```python
#диаграмма рассеивания
# sns.scatterplot(x='year', y='selling_price', data=cars)
sns.scatterplot(x='seats', y='mileage', hue='fuel', data=cars)
```

```python
#диаграмма рассеивания лин регр
sns.regplot(x='seats', y='mileage', data=cars)
```

```python
sns.relplot(x='seats', y='mileage', data=cars, kind='scatter', col='transmission', col_wrap=2, hue='fuel')
plt.show()
plt.show()
```

```python
#диаграмма рассеивания лин регр
sns.lmplot(x='seats', y='mileage', data=cars, col='transmission', col_wrap=2, hue='fuel')
```

```python
#лин график
sns.lineplot(x='seats', y='mileage', data=cars)
plt.show()
```

```python
#сводные диаграммы
sns.jointplot(x='year', y='selling_price', data=cars, kind='kde')
plt.show()
```

```python
sns.jointplot(x='year', y='selling_price', data=cars, hue='transmission')
plt.show()
```
