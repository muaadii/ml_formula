```python
import numpy as np
import matplotlib.pyplot as plt
```

# График №1

```python
x = np.linspace(1, 20, 5)

seed = np.random.seed(6)

y1 = np.random.randint(1, 20, (5))
y2 = np.random.randint(1, 20, (5))

plt.plot(x, y1, 'o-r', label='line 1')
plt.plot(x, y2, 'o--g', label='line 2')
plt.legend()
plt.show()
```

```python
grid = plt.GridSpec(2, 2)
```

```python
fig = plt.figure(figsize=(12, 8))


plot1 = plt.subplot(grid[0, :])
plot2 = plt.subplot(grid[1, 0])
plot3 = plt.subplot(grid[1, 1])

x = np.linspace(0, 20, 5)
seed = np.random.seed(6)

y1 = np.random.randint(0, 20, (5))
y2 = (x - 10) ** 2
y3 = - (x - 10) ** 2 

plot1.plot(x, y1)
plot2.plot(x, y2)
plot3.plot(x, y3)


plt.show()
```

# График №3

```python
x = np.linspace(-20, 20, 100)

plt.plot(x, x ** 2)
plt.annotate(
    'min', 
    xy=(0, 0),
    xytext=(0, 100),        
    fontsize=14,
    ha='center',
    arrowprops=dict(
        facecolor='green',
        lw=1,
    )
)
plt.show()
```

# График №4

```python
seed = np.random.seed(6)
xy = np.random.randint(0, 100, (6, 10))

fig, ax = plt.subplots(figsize=(8, 4))
p = ax.imshow(xy, cmap='viridis')
ax.figure.colorbar(p, ax=ax, shrink=0.4, aspect=8)
plt.show()
```

# График №5

```python
x = np.linspace(0, 15, 300)
y = np.cos(x)

fig, ax = plt.subplots(figsize=(8, 4))

ax.plot(x, y, color='red')
ax.fill_between(x, y, 0)
plt.show()
```

# График №6

```python
x = np.linspace(0, 15, 300)
y = np.cos(x)

y[y < -0.5] = np.nan

fig, ax = plt.subplots(figsize=(8, 4))

ax.plot(x, y)
plt.ylim((-1, 1))
```

# График №7

```python
fig, ax = plt.subplots(1, 3, figsize=(12, 4))

y = x = np.arange(1, 10)

ax[0].step(x, y, where='post', marker='o', color='green')
ax[0].grid()

ax[1].step(x, y, where='pre', marker='o', color='green')
ax[1].grid()

ax[2].step(x, y, where='mid', marker='o', color='green')
ax[2].grid()

plt.show()
```

# График №8

```python
x = np.linspace(0, 10, 50)

y1 =  x * (10 - x)
y2 = 2 * x * (10 - x)
y3 = 2 * x * (15 - x)

plt.stackplot(x, y1,  y2,  y3, labels=['y1',  'y2',  'y3'])
plt.legend()
plt.show()
```

# График №8

```python
seed = np.random.seed(6)
x = np.random.rand(5)
x = x / x.sum() *  100

labels = ['Toyota', 'Ford', 'Jaguar', 'AUDI', 'BMV']

fig, ax = plt.subplots(figsize=(7, 7))

ax.pie(x, labels=labels, explode=[0.1, 0, 0, 0, 0])
plt.show()
```

# График №9

```python
seed = np.random.seed(6)
x = np.random.rand(5)
x = x / x.sum() *  100

labels = ['Toyota', 'Ford', 'Jaguar', 'AUDI', 'BMV']

fig, ax = plt.subplots(figsize=(7, 7))

ax.pie(x, labels=labels, wedgeprops={'width': 0.5})
plt.show()
```
