```python
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
X = iris.data[:, :4]
y = iris.target[:]
```

```python
X = X[y != 2]
y = y[y != 2]
```

```python
pca = PCA(n_components=2)
new_X = pca.fit_transform(X)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y)
plt.show()
```
