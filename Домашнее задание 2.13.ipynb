```python
from sklearn.datasets import fetch_olivetti_faces
faces = fetch_olivetti_faces().images

fig, ax = plt.subplots(7,7)
fig.subplots_adjust(hspace=0.5, wspace=0.3)
for i in range(7):
    for j in range(7):
        ax[i,j].xaxis.set_major_formatter(plt.NullFormatter())
        ax[i,j].yaxis.set_major_formatter(plt.NullFormatter())
        ax[i, j].imshow(faces[i*7+j], cmap='binary_r')
plt.show()
```

```python
def ff(value, tick_number):
    N = int(np.round(2 * value / np.pi))
    if N == 0:
        return 0
    elif N == 1:
        return r'$\frac{\pi}{2}$'
    elif N == 2:
        return r'$\pi$'
    elif N%2 == 0:
        t = int(N/2)
        return f'{t}' + r'$\pi$'
    else:
        return f'{N}' + r'$\frac{\pi}{2}$'
    return value

x = np.linspace(0,4*np.pi ,1000)

fig, ax = plt.subplots()
ax.plot(x, np.sin(x), label='sin(x)')
ax.plot(x, np.cos(x), label='cos(x)')

ax.legend()
ax.grid(True)
ax.set_xlim(0,4*np.pi )

ax.xaxis.set_major_locator(plt.MultipleLocator(np.pi/2))
ax.xaxis.set_minor_locator(plt.MultipleLocator(np.pi/4))

ax.xaxis.set_major_formatter(plt.FuncFormatter(ff))
plt.show()
```

```python
fig,ax = plt.subplots(8,1)


x = np.linspace(0,10,10)

for i in ax.flat:
    i.plot(x,x*0+2)

ax[0].xaxis.set_major_locator(plt.NullLocator())
ax[1].xaxis.set_major_locator(plt.MultipleLocator(0.8))
ax[2].xaxis.set_major_locator(plt.FixedLocator(np.linspace(1,9,4,9)))
ax[3].xaxis.set_major_locator(plt.LinearLocator(numticks=7))
ax[4].xaxis.set_major_locator(plt.IndexLocator(base=1.5, offset=0.3))
ax[5].xaxis.set_major_locator(plt.AutoLocator())
ax[6].xaxis.set_major_locator(plt.MaxNLocator(8))
ax[7].xaxis.set_major_locator(plt.LogLocator(base=3, numticks=8))

import matplotlib.ticker as mtick

ax[1].xaxis.set_major_formatter(plt.NullFormatter())
ax[3].xaxis.set_major_formatter(plt.FormatStrFormatter('%.2f $m^2$'))
# ax[2].xaxis.set_major_formatter(plt.FixedFormatter('a','b','c','d'))
ax[4].xaxis.set_major_formatter(mtick.PercentFormatter(xmax=4))

plt.subplots_adjust(hspace=1)
plt.show()
```

```python
from mpl_toolkits import mplot3d
def f(x,y):
    return np.sin(np.sqrt(x**2+y**2))

x = np.linspace(-6,6,30)
y = np.linspace(-10,10,50)
# print(x.shape)
# print(y.shape)

X, Y = np.meshgrid(x,y)
# print(X.shape)
# print(Y.shape)

Z = f(X,Y)
# print(Z.shape)
# print(Z)

fig,ax = plt.figure(), plt.axes(projection='3d')
ax.contour3D(X,Y,Z,100)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

ax.view_init(20,0)

plt.show()
```
